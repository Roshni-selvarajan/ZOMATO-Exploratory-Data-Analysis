```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0: 'Rating count'})
ratings.head()
```

## Observation
1. **When the rating is lie between 4.5 - 4.9, the rating is Exce**llent
2. **When the rating is lie between 4.0 - 4.4, the rating is Very Good**
3. **When the rating is lie between 3.5 - 3.9, the rating is Good**
4. **When the rating is lie between 2.5 - 3.4, the rating is Average**
5. **When the rating is lie between 1.8 - 2.4, the rating is Poor**

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating count',data=ratings,palette='viridis')
```

```python
sns.barplot(x='Aggregate rating',y='Rating count',hue='Rating color',data=ratings,palette=['white','red','orange','yellow','green','green'])
```

# Observations    
1.Not rated count is very high  
2.Maximum no of rating are between 2.5 to 3.4

```python
#count plot
sns.countplot(x='Rating color',data=ratings,palette=['white','red','orange','yellow','green','green'])
```
